# IMPORTING THE LIBRARIES AND LOADING THE DATASET

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Loan.csv')
```

```python
df.head(30)
```

```python
df.shape
```

# DATA CLEANING AND FINDING THE MISSING VALUES

```python
df.isnull()
# gives result as true or false
```

```python
# finding total null values in each column
total = df.isnull().sum
```

```python
total()
```

# working on the columns with null values
# what are more in number we will replace the null values by that which is higher in number

```python
df['Gender'].value_counts()
```

# males are much larger in number than females , so we will fill the null values in Gender column as Males.

```python
df.Gender = df.Gender.fillna('Male')
```

```python
df.head(30)
```

```python
df['Married'].value_counts()
```

```python
df.Married = df.Married.fillna('Yes')
```

```python
df['Dependents'].value_counts()
```

```python
df.Dependents = df.Dependents.fillna('0')
```

```python
df['Self_Employed'].value_counts()
```

```python
df.Self_Employed = df.Self_Employed.fillna('No')
```

```python
df['LoanAmount'].value_counts()
```

```python
df.LoanAmount = df.LoanAmount.fillna(df.LoanAmount.mean())
```

```python
df.head(30)
```

```python
df['Loan_Amount_Term'].value_counts()
```

```python
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(360.0)
```

```python
df.head(30)
```

```python
df['Credit_History'].value_counts()
```

```python
df.Credit_History = df.Credit_History.fillna('1.0')
```

```python
df.head(30)
```

```python
total = df.isnull().sum
total()
```

```python
df.head(50)
```

# CONVERTING CATEGORICAL VALUE INTO NUMERIC USING LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status','Dependents']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
    
```

```python
df.head(50)
```

# FEATURE SELECTION
# iloc indexer for Pandas data frame is used for integer-location based indexing by position.

```python
x = df.iloc[:,1:12]
y = df.iloc[:,12]
```

# x is the independent variable

```python
x.head()
```

# y is the dependent variable

```python
y.head()
```

# Splitting The Data

# here we will split the data in 80:20 ratio. 80% for training the model and 20% for testing the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.20, random_state = 0)
```

```python
x_test
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

# BUILDING DECISION TREE MODEL:

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_predictions = model.predict(x_test)
y_predictions
```

```python
y_test
```

# MODEL EVALUATION

```python
# import the metrics class
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_predictions)
cnf_matrix
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_predictions))
```

# Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train, y_train)
```

```python
y_pred = logistic_regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

# Model Evaluation Using Confusion Matrix

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
